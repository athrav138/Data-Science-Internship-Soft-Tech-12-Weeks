```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
# Collecting the dat from CSV File

df = pd.read_csv("vgsales.csv")
```

```python
df["Name"] = df["Name"].astype('string')
df["Platform"] = df["Platform"].astype('string')
df["Genre"] = df["Genre"].astype('string')
df["Publisher"] = df["Publisher"].astype('string')
```

```python
df[['Year', 'Publisher']].isnull().all(axis=1).sum()
```

```python
df = df.dropna(
    subset=['Year', 'Publisher'],
    how='all'
)
```

```python
df[['Year', 'Publisher']].isnull().sum()
```

```python
df["Year"] = df["Year"].astype(int)
```

```python
df["Year"] = pd.to_numeric(df["Year"])
```

```python
df['Year'] = df['Year'].fillna(
    df['Year'].median()
)
```

```python
df = df.dropna(subset=['Publisher'])
```

```python
df.to_csv("cleaned_vgsales.csv",index=False)
```
