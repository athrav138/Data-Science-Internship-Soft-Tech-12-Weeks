```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data  = {
    "Name":["Atharv","Shravan","Om","Shree","Tom"],
    "Age":[25,45,34,38,30],
    "Salary":[25000,45000,34000,38000,30000],
    "City":["Sangli","Pune","Miraj","Jaysingpur","Pune"]
}

df = pd.DataFrame(data)
print(df)
```

```python
df["Age_Group"] = pd.cut(df["Age"],
                        bins = [18,30,40,60],
                         labels = ['Fresher','Junior','Senior'])
print(df)
```

```python
def salary_categories(sal):
  if sal<=30000:
    return "Low"
  elif sal<=40000:
    return "Medium"
  else:
    return "High"

df["Salary_Category"] = df["Salary"].apply(salary_categories)
df
```

```python
city_encoded = pd.get_dummies(df["City"])
print(city_encoded)
```

```python
df["Combined"] = df["Salary"]/df["Age"]
print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler

scalar = MinMaxScaler()
df["Salary_scale"] = scalar.fit_transform(df[["Salary"]])
df["Age_scale"] = scalar.fit_transform(df[["Age"]])
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df["Salary_scale"] = scaler.fit_transform(df[["Salary"]])
df["Age_scale"] = scaler.fit_transform(df[["Age"]])
df
```
