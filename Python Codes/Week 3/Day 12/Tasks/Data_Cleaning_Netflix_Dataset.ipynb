```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('netflix_titles.csv')

df.head()
```

```python
print("Shape:", df.shape)

df.info()
```

```python
df.isnull().sum()
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

df.columns
```

```python
missing_percent = round((df.isnull().sum()/len(df))*100,2)

missing_percent.sort_values(ascending=False)
```

```python
df['director'] = df['director'].fillna('Not Available')
```

```python
df['cast'] = df['cast'].fillna('Not Available')
```

```python
df['country'] = df['country'].fillna(df['country'].mode()[0])
```

```python
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
```

```python
df['date_added'] = df['date_added'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df['duration'] = df['duration'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df['date_added'] = pd.to_datetime(
    df['date_added'],
    errors='coerce'
)

df['date_added'].head()
```

```python
df['year_added'] = df['date_added'].dt.year

df['month_added'] = df['date_added'].dt.month
```

```python
df.dtypes
```

```python
df[df['release_year'] > pd.Timestamp.now().year]
```

```python
df['type'].unique()
```

```python
df['type'] = df['type'].str.title()
```

```python
df['date_added'] = df['date_added'].fillna("2000-01-01")
```

```python
df['year_added'] = df['year_added'].fillna("2000-01-01")
df['month_added'] = df['month_added'].fillna("2000-01-01")

df.isnull().sum()
```
