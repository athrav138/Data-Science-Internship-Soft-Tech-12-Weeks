```python
df["Joining_Date"] = pd.to_datetime(df["Joining_Date"],errors="coerce")
df["Joining_Date"] = df["Joining_Date"].dt.strftime("%Y-%m-%d")
```

```python

def parse_joining_date(date_str):
    if pd.isna(date_str):
        return pd.NaT

    s = str(date_str).strip()
    if not s or s.lower() == 'nan':
        return pd.NaT

    # List of explicit formats to try
    formats = [
        "%m/%d/%Y",  # e.g., 12/2/2026, 1/11/2025
        "%d-%b-%y",  # e.g., 20-Mar-25, 15-Jan-24
        "%d-%m-%y",  # e.g., 31-12-22, 30-11-21
        "%Y.%m.%d",  # e.g., 2024.06.15
        "%B %d %Y",  # e.g., June 10 2023
        "%b %d %Y",  # e.g., Aug 15 2023
    ]

    for fmt in formats:
        try:
            return pd.to_datetime(s, format=fmt)
        except ValueError:
            continue

    # Fallback: let pandas try to infer without a specific format
    try:
        return pd.to_datetime(s)
    except ValueError:
        return pd.NaT # Return NaT if all attempts fail

# Apply the robust parsing function to the 'Joining_Date' column
df["Joining_Date"] = df["Joining_Date"].apply(parse_joining_date)

# Now df["Joining_Date"] is of datetime64[ns] dtype with NaT for unparseable/missing dates
# We will display the column and its dtype to confirm the change
display(df["Joining_Date"])
display(df["Joining_Date"].dtype)
```

```python
df['Joining_Date'] = df['Joining_Date'].fillna('2026-06-01')
```

```python
median_age = df.loc[(df["Age"] > 0) & (df["Age"] < 125), "Age"].median()

df["Age"] = df["Age"].where(
    (df["Age"] > 0) & (df["Age"] < 125),
    median_age
)
```

```python
df['Age']=df["Age"].astype(int)
```

```python
df['Gender'] = df['Gender'].str.title().replace({'M':'Male','F':'Female'})
df['Department'] = df["Department"].str.upper().replace({"INFORMATION TECHNOLOGY":'IT','I.T.':'IT','FINANCE':'FIN','UNKNOWN':'FIN'})
```

```python
df.dropna(thresh=2,inplace=True)
```

```python
df
```

```python
df['Salary'] = df['Salary'].astype(str).str.replace('?', '', regex=False)
df['Salary'] = df['Salary'].str.replace('ance', '', regex=False)
df['Salary'] = df['Salary'].str.replace(',', '', regex=False)
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
```

```python
df = df.astype('object')
df.loc[:, "Country"] = "India"


```

```python
df
```

```python
def clean_phone_number(phone):
    if pd.isna(phone) or phone == 'NA':
        return 'NA'
    phone_str = str(phone)
    # Remove non-numeric characters
    cleaned_phone = ''.join(filter(str.isdigit, phone_str))

    # If the cleaned number is 10 digits long, return it
    if len(cleaned_phone) == 10:
        return cleaned_phone
    # Otherwise, return 'NA'
    return 'NA'

df['Phone'] = df['Phone'].apply(clean_phone_number)
```

```python
# Convert 'Marks' to numeric, coercing errors to NaN
df['Marks'] = pd.to_numeric(df['Marks'], errors='coerce')

# Calculate the mean of valid marks (between 1 and 100)
valid_marks = df.loc[(df['Marks'] >= 1) & (df['Marks'] <= 100), 'Marks']
mean_valid_marks = valid_marks.mean()

# Replace invalid marks (outside 1-100) and NaN with the mean of valid marks
df['Marks'] = df['Marks'].mask((df['Marks'] < 1) | (df['Marks'] > 100) | df['Marks'].isna(), mean_valid_marks)

# Convert 'Marks' to integer type (if appropriate, or keep as float if there are non-integer means)
df['Marks'] = df['Marks'].astype(int)

display(df['Marks'])
```
