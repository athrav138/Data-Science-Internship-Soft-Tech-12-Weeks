### Final Cleanliness Check

```python
# Check for any remaining missing values
missing_values_after_cleaning = df.isnull().sum()
print("Missing values after cleaning:")
print(missing_values_after_cleaning[missing_values_after_cleaning > 0])

if missing_values_after_cleaning.sum() == 0:
    print("\nAll missing values have been handled.")
else:
    print("\nThere are still missing values in the DataFrame.")
```

```python
# Display data types and non-null counts to assess column readiness
df.info()
```

Based on the checks, if no missing values are reported and data types are as expected, the DataFrame should be ready for analysis. Some columns like 'Name', 'Ticket', and 'Fare' might still need specific handling depending on the type of analysis or modeling intended, but for general cleanliness, it should be good.

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df["Name"].str.strip()
```

```python
# Identify object (string) columns
object_cols = df.select_dtypes(include='object').columns

# Strip whitespace from each object column
for col in object_cols:
    df[col] = df[col].astype(str).str.strip()

print("Whitespace stripped from the following columns:")
print(object_cols)

# Display a sample of the DataFrame to show changes
display(df.head())
```

```python
# Identify rows where 'Age' is outside the typical human age range (e.g., less than 1 or greater than 125)
invalid_age_rows = df[(df['Age'] < 1) | (df['Age'] > 125)]

if not invalid_age_rows.empty:
    print("Rows with potentially invalid 'Age' values:")
    display(invalid_age_rows)
else:
    print("No 'Age' values found outside the range of 1 to 125.")
```
