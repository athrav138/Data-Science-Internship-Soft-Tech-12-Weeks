```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df = pd.read_csv("cleaned_vgsales.csv")
```

## Feature Engineering Summary

Here's a summary of the feature engineering steps performed:

1.  **Decade_group**: Created a categorical feature from the `Decade` column, grouping years into decades (e.g., '80s', '90s'). This involved careful binning with `right=False` in `pd.cut` to correctly assign boundary years and extending the bins to include 2020.
2.  **Calculated_Sales**: Summed regional sales (`NA_Sales`, `EU_Sales`, `JP_Sales`, `Other_Sales`) to get total sales, which was then compared with `Global_Sales`.
3.  **Top_Region**: Identified the region with the highest sales for each game among `NA_Sales`, `EU_Sales`, `JP_Sales`, and `Other_Sales`.
4.  **Sales_Category**: Categorized `Global_Sales` into 'Low', 'Medium', 'High', and 'Blockbuster' bins.
5.  **Game_Age**: Calculated the age of each game by subtracting its `Year` from 2026.
6.  **NA_Percentage**: Calculated North American sales as a percentage of `Global_Sales`.
7.  **EU_Percentage**: Calculated European sales as a percentage of `Global_Sales`.

```python
df
```

```python
df["Decade"] = (df["Year"] // 10) * 10
df
```

```python
df["Decade_group"] = pd.cut(df["Decade"], bins=[1980, 1990, 2000, 2010, 2020, 2030], labels=["80s", "90s", "00s", "10s", "20s"], right=False)

df = df.drop(columns=['Decade_group_filled'], errors='ignore')

df
```

```python
df
```

```python
df["Calculated_Sales"] = (
    df["NA_Sales"]
    + df["EU_Sales"]
    + df["JP_Sales"]
    + df["Other_Sales"]
)
```

```python
df
```

```python
df[["Global_Sales", "Calculated_Sales"]].head()
```

```python
df["Top_Region"] = df[
    ["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]
].idxmax(axis=1)
```

```python
df
```

```python
df["Top_Region"].value_counts()
```

```python
df["Sales_Category"] = pd.cut(
    df["Global_Sales"],
    bins=[0, 1, 10, 50, 100],
    labels=["Low", "Medium", "High", "Blockbuster"]
)
```

```python
df
```

```python
df["Game_Age"] = 2026 - df["Year"]
```

```python
df
```

```python
df["Game_Age"].describe()
```

```python
df["NA_Percentage"] = (
    df["NA_Sales"] / df["Global_Sales"]
) * 100
```

```python
df["EU_Percentage"] = (
    df["EU_Sales"] / df["Global_Sales"]
) * 100
```

```python
df
```

Feature Scaling

```python
scaler = MinMaxScaler()
df["Global_Sales_Scaler"] = scaler.fit_transform(df[["Global_Sales"]])
df["NA_Percentage_Scaler"] = scaler.fit_transform(df[["NA_Percentage"]])
df["EU_Percentage_Scaler"] = scaler.fit_transform(df[["EU_Percentage"]])
df
```

```python
scaler = StandardScaler()
df["Global_Sales_Standard_Scaler"] = scaler.fit_transform(df[["Global_Sales"]])
df["NA_Percentage_Standard_Scaler"] = scaler.fit_transform(df[["NA_Percentage"]])
df["EU_Percentage_Standard_Scaler"] = scaler.fit_transform(df[["EU_Percentage"]])
df
```

```python
df.to_csv("feature_vgsales.csv",index=False)
```
